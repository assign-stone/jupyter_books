```python
def vowel_counter(word):
    ctr = 0
    vowels = ['a','e','i','o','u']
    for character in word.lower():
        if character in vowels:
            ctr+=1
    return ctr
```

```python
# groupby
# Sample dataset
data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Finance', 'Finance'],
    'Gender': ['F', 'M', 'F', 'M', 'F', 'M'],
    'Salary': [50000, 55000, 60000, 65000, 52000, 58000],
    'Bonus': [5000, 4000, 6000, 5000, 4500, 4700]
}

df = pd.DataFrame(data)
df
```

```python

# Average salary per department
df.groupby('Department')['Salary'].mean().reset_index()
```

```python

# Salary stats per department
df.groupby('Department')['Salary'].agg(['mean', 'min', 'max']).reset_index()
```

```python
# Different aggregations for salary and bonus
df.groupby('Department').agg({
    'Salary': 'mean',
    'Bonus': 'mean'
}).reset_index()
```

```python
# Group by per Department per Gender average salary
df.groupby(['Department', 'Gender'])['Salary'].mean().reset_index()
```

```python

# Named aggregations with new column names
agg_df = df.groupby('Department').agg(
    Avg_Salary=('Salary', 'mean'),
    Total_Bonus=('Bonus', 'sum')
)
agg_df.reset_index()
```

```python

# Named aggregations with new column names
agg_df = df.groupby('Department').agg(
    min_salary=('Salary', 'min'),
    max_salary=('Salary', 'max')
)
agg_df.reset_index()
```

```python


# Calculate range (max - min) of salary per department
salary_range = df.groupby('Department')['Salary'].agg(lambda x: x.max() - x.min())
salary_range.reset_index()
```

```python
# merging and join
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Score': [85, 90, 95]
})
```

```python
df1
```

```python
df2
```

```python

# Merge on common column
pd.merge(df1, df2, on='ID', how='inner')
```

```python

# Merge on common column
outer_join = pd.merge(df1, df2, on='ID', how='outer')
```
