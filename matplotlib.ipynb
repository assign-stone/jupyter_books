```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [2,3,4,5]
y = np.array
x = np.linspace(0,5,5)
y = x**2
```

```python
plt.plot(x,y,'r')
plt.xlabel('Sugar intake')
plt.ylabel('Diabeties')
plt.title('Diabeties vs Sugar Intake')
plt.show()
```

```python
# With placement and sizing attributes

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])#front two parameters of placement and last two is for width and height respectively.
axes.plot(x,y,'b')
axes.set_xlabel('Sugar intake')
axes.set_ylabel('Diabeties')
axes.set_title('Diabeties vs Sugar Intake')
```

```python
#figure in figure

fig2 = plt.figure()

axes1 = fig2.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig2.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'b')

plt.xlabel('Sugar intake')
plt.ylabel('Diabeties')
plt.title('Diabeties vs Sugar Intake')

axes2.plot(y,x,'r')
plt.xlabel('Sugar intake')
plt.ylabel('Diabeties')
plt.title('Diabeties vs Sugar Intake')
```

```python
#resizing of figure

fig,axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x,y,'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('my fig')
fig
plt.tight_layout()
```

```python
fig,axes = plt.subplots(figsize = (12,3))


axes.plot(x,y,'b')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
#save the figure
fig.savefig("fig.jpg",dpi=200)
```

```python
#legends

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend(loc=1)
```
